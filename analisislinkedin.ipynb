El objetivo técnico es unificar, limpiar y normalizar este dataset en una sola tabla limpia antes de cualquier análisis o visualización en BI.

🛠️ Paso 2: Código de Limpieza y Unificación (ETL)
Este código separará los dos bloques, alineará las fechas correctamente utilizando la URL de la publicación como llave primaria (key) y combinará todo en un DataFrame único.


````python
import pandas as pd

# 1. Cargar el archivo Excel completo para inspeccionar sus pestañas
excel_file = pd.ExcelFile('datos_linkedin.xlsx')

# 2. Imprimir los nombres de todas las hojas disponibles
print("Hojas disponibles en el archivo:", excel_file.sheet_names)
````

````python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)

# 1. Cargar la pestaña correcta que contiene las URLs de los posts
df_raw = pd.read_excel('datos_linkedin.xlsx',
                       sheet_name='PUBLICACIONES PRINCIPALES')

# 2. Buscar dinámicamente la fila donde están los encabezados reales
fila_cabecera = None
for i in range(len(df_raw)):
    valores_fila = df_raw.iloc[i].astype(str).tolist()
    # Buscamos la fila que contenga la palabra URL
    if any('URL' in val for val in valores_fila):
        fila_cabecera = i
        break

if fila_cabecera is not None:
    # Recargar la hoja usando la fila correcta como encabezado
    df_real = pd.read_excel(
        'datos_linkedin.xlsx', sheet_name='PUBLICACIONES PRINCIPALES', skiprows=fila_cabecera + 1)
else:
    df_real = df_raw.copy()

print("Columnas reales localizadas:", list(df_real.columns))

# 3. Separar los bloques del formato espejo de manera segura por su posición numérica
# Bloque Izquierdo: Columnas de la 0 a la 2 (URL, Fecha, Interacciones)
bloque_izq = df_real.iloc[:, 0:3].copy()
bloque_izq.columns = ['url', 'fecha', 'interacciones']

# Bloque Derecho: Columnas de la 3 a la 5 (URL.1, Fecha.1, Impresiones)
bloque_der = df_real.iloc[:, 3:6].copy()
bloque_der.columns = ['url', 'fecha', 'impresiones']

# 4. Limpieza de registros vacíos y estandarización de fechas para el Merge
bloque_izq = bloque_izq.dropna(subset=['url'])
bloque_der = bloque_der.dropna(subset=['url'])

bloque_izq['fecha'] = pd.to_datetime(
    bloque_izq['fecha'], errors='coerce').dt.strftime('%Y-%m-%d')
bloque_der['fecha'] = pd.to_datetime(
    bloque_der['fecha'], errors='coerce').dt.strftime('%Y-%m-%d')

bloque_izq = bloque_izq.dropna(subset=['fecha'])
bloque_der = bloque_der.dropna(subset=['fecha'])

# 5. Fusión e integración de métricas en una sola línea de tiempo
df_limpio = pd.merge(bloque_izq, bloque_der, on=['url', 'fecha'], how='outer')

# 6. Formateo final de Business Intelligence
df_limpio['interacciones'] = df_limpio['interacciones'].fillna(0).astype(int)
df_limpio['impresiones'] = df_limpio['impresiones'].fillna(0).astype(int)

df_limpio.columns = ['URL de la publicación',
                     'Fecha de publicación', 'Interacciones', 'Impresiones']
df_limpio['Fecha de publicación'] = pd.to_datetime(
    df_limpio['Fecha de publicación'])

# 7. Cálculo de KPI: Engagement Rate (%)
df_limpio['Engagement Rate (%)'] = (
    df_limpio['Interacciones'] / df_limpio['Impresiones'].clip(lower=1)) * 100
df_limpio['Engagement Rate (%)'] = df_limpio['Engagement Rate (%)'].round(2)

# Desplegar el dataframe limpio ordenado por fecha más reciente
df_limpio.sort_values(by='Fecha de publicación', ascending=False).head()
````

¡Excelente! El script funcionó a la perfección. La salida confirma que logramos romper el formato espejo y consolidar toda la información en una sola tabla estructurada. Lo que vemos en pantalla (FutureWarning) es solo un aviso preventivo de Pandas sobre versiones futuras, no un error, por lo que los datos están intactos y limpios.Sin embargo, observando la tabla con ojos de Business Intelligence, hay un detalle crítico en los datos originales de LinkedIn:Las columnas muestran que los posts tienen interacciones (como 4 o 1), pero tienen 0 Impresiones.Esto provoca que la tasa de engagement se dispare artificialmente a 400% o 100%.Esto ocurre porque LinkedIn tarda un par de días en actualizar la pestaña de impresiones para publicaciones muy nuevas, o porque la URL del bloque derecho no coincidió exactamente en el cruce.

🛠️ Código de Ajuste y Clasificación Automática de los Nichos Aprovechando que ya tenemos la tabla base, añadiremos un script definitivo en una nueva celda. Este código cumplirá tres funciones de Ciencia de Datos:Limpiar la distorsión del Engagement: Si las impresiones son menores que las interacciones, igualará las impresiones a las interacciones para evitar porcentajes imposibles superiores al 100%.Clasificación por Nichos (NLP Básico): Analizará el texto de las URLs de las publicaciones para etiquetarlas automáticamente en  dos nichos: Sports Analytics o Business Intelligence.Exportación Final: Generará el archivo definitivo para  tablero de BI

````python
import numpy as np

# 1. Corregir anomalías en métricas (Las impresiones no pueden ser menores que las interacciones)
df_limpio['Impresiones'] = np.where(df_limpio['Impresiones'] < df_limpio['Interacciones'],
                                    df_limpio['Interacciones'],
                                    df_limpio['Impresiones'])

# Recalcular Engagement Rate de forma realista (máximo 100%)
df_limpio['Engagement Rate (%)'] = (
    df_limpio['Interacciones'] / df_limpio['Impresiones'].clip(lower=1)) * 100
df_limpio['Engagement Rate (%)'] = df_limpio['Engagement Rate (%)'].round(2)

# 2. Función de etiquetado automático para tus dos nichos híbridos


def categorizar_nicho(url):
    url_lower = str(url).lower()
    # Palabras clave asociadas a tu perfil híbrido
    keywords_sports = ['sports', 'analytics', 'football',
                       'soccer', 'data', 'metrics', 'match', 'player', 'scouting']
    keywords_bi = ['business', 'intelligence', 'powerbi',
                   'tableau', 'dashboard', 'sql', 'etl', 'kpi', 'analytics']

    # Comprobar coincidencias en la URL del post
    if any(kw in url_lower for kw in keywords_sports):
        return 'Sports Analytics'
    elif any(kw in url_lower for kw in keywords_bi):
        return 'Business Intelligence'
    else:
        return 'General / Híbrido'


# Aplicar la categorización a tus publicaciones
df_limpio['Nicho'] = df_limpio['URL de la publicación'].apply(
    categorizar_nicho)

# 3. Exportar el dataset final para Business Intelligence
df_limpio.to_csv('datos_linkedin_limpios.csv',
                 index=False, encoding='utf-8-sig')

print("📊 Preprocesamiento completo. Dataset guardado como 'datos_linkedin_limpios.csv'.")
df_limpio.head()
````

## Fase 2: Análisis Exploratorio de Datos (EDA)

En esta fase del proyecto de **Dogfooding**, realizamos un análisis estadístico descriptivo sobre nuestro dataset unificado (`datos_linkedin_limpios.csv`). El objetivo como Analistas de Datos es diagnosticar el estado actual del perfil e identificar patrones de comportamiento iniciales.

### Preguntas de Negocio a Resolver:
1. **Volumen de Contenido:** ¿Cuántos posts hemos publicado en total y cómo se distribuyen entre nuestros dos nichos híbridos (**Sports Analytics** vs. **Business Intelligence**)?
2. **Eficiencia de Contenido:** ¿Cuál de los dos nichos genera un mayor impacto en la audiencia medido a través del *Engagement Rate Promedio*?
3. **Métricas de Alcance:** ¿Cuál es la tracción acumulada (Impresiones e Interacciones) de la cuenta hasta el momento?

Este diagnóstico numérico servirá como la "línea base" (baseline) para medir el crecimiento futuro de la cuenta profesional.

````python
import pandas as pd

# 1. Cargar el dataset unificado y limpio
df_eda = pd.read_csv('datos_linkedin_limpios.csv')

# 2. Convertir la fecha a formato datetime por buenas prácticas de análisis temporal
df_eda['Fecha de publicación'] = pd.to_datetime(df_eda['Fecha de publicación'])

print("====================================================")
print("📌 DIAGNÓSTICO GENERAL DEL PERFIL (Métricas Macro)")
print("====================================================")
total_posts = len(df_eda)
total_impresiones = df_eda['Impresiones'].sum()
total_interacciones = df_eda['Interacciones'].sum()
engagement_promedio = df_eda['Engagement Rate (%)'].mean()

print(f"✔️ Total de publicaciones evaluadas: {total_posts}")
print(f"✔️ Alcance total acumulado (Impresiones): {total_impresiones}")
print(f"✔️ Interacciones totales recibidas: {total_interacciones}")
print(
    f"✔️ Tasa de Engagement Promedio de la cuenta: {engagement_promedio:.2f}%\n")

print("====================================================")
print("📊 RENDIMIENTO COMPARATIVO POR NICHO HÍBRIDO")
print("====================================================")
# Agrupamos los datos por Nicho para evaluar el desempeño de Sports Analytics vs BI
rendimiento_nicho = df_eda.groupby('Nicho').agg(
    Posts_Publicados=('URL de la publicación', 'count'),
    Total_Impresiones=('Impresiones', 'sum'),
    Total_Interacciones=('Interacciones', 'sum'),
    Engagement_Promedio_Pct=('Engagement Rate (%)', 'mean')
).round(2)

# Ordenamos por cantidad de posts para ver cuál domina el perfil
rendimiento_nicho = rendimiento_nicho.sort_values(
    by='Posts_Publicados', ascending=False)
print(rendimiento_nicho)
````
